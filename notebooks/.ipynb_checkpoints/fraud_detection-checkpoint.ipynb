```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset
data = pd.read_csv(r'D:\My Projects\Credit_Card_Fraud_Detection\data\creditcard.csv')
```

```python
# Extracting basic information
print("✅ Dataset Loaded Successfully!")
print(data.shape)
data.head()
```

```python
# Checking class distribution
print("\nClass Distribution:")
print(data['Class'].value_counts())
```

```python
# Visualizing class distribution
plt.figure(figsize=(6,4))
sns.countplot(data=data, x='Class')
plt.title('Class Distribution (0 = No Fraud, 1 = Fraud)')
plt.show()
```

```python
# Preprocessing: Scaling 'Time' and 'Amount'
from sklearn.preprocessing import StandardScaler

# Create new scaled features
data['scaled_amount'] = StandardScaler().fit_transform(data['Amount'].values.reshape(-1, 1))
data['scaled_time'] = StandardScaler().fit_transform(data['Time'].values.reshape(-1, 1))

# Dropping original 'Amount' and 'Time' features
data = data.drop(['Amount', 'Time'], axis=1)

# Reordering columns for neatness
scaled_amount = data['scaled_amount']
scaled_time = data['scaled_time']
data.drop(['scaled_amount', 'scaled_time'], axis=1, inplace=True)
data.insert(0, 'scaled_amount', scaled_amount)
data.insert(1, 'scaled_time', scaled_time)

print("✅ Scaling Done. Data ready for next steps!")
data.head()
```

```python
# Dealing with Imbalance using SMOTE
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

# Separating features and target
X = data.drop('Class', axis=1)
y = data['Class']

# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print(f"Shape before SMOTE:")
print(f"X_train: {X_train.shape}, y_train: {y_train.shape}")
print(f"Fraudulent transactions in y_train before SMOTE:\n{y_train.value_counts()}")

# Applying SMOTE
sm = SMOTE(random_state=42)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)

print(f"\n✅ After applying SMOTE:")
print(f"X_train_res: {X_train_res.shape}, y_train_res: {y_train_res.shape}")
print(f"Fraudulent transactions in y_train after SMOTE:\n{y_train_res.value_counts()}")
```

```python
# Training XGBoost Classifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve, accuracy_score

# Instantiate the model
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)

# Training on the resampled training data
xgb_model.fit(X_train_res, y_train_res)

# Making predictions on test data
y_pred = xgb_model.predict(X_test)

# Evaluation Metrics
print("\n✅ Classification Report:")
print(classification_report(y_test, y_pred))

print("\n✅ Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\n✅ Accuracy Score:")
print(accuracy_score(y_test, y_pred))

print("\n✅ ROC-AUC Score:")
print(roc_auc_score(y_test, y_pred))

# Plotting ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, xgb_model.predict_proba(X_test)[:,1])

plt.figure(figsize=(8,6))
plt.plot(fpr, tpr, label=f'XGBoost (AUC = {roc_auc_score(y_test, y_pred):.2f})')
plt.plot([0,1], [0,1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

# Training Random Forest
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train_res, y_train_res)

# Predicting
y_pred_rf = rf_model.predict(X_test)

# Evaluating
print("\n✅ Random Forest Evaluation:")
print(classification_report(y_test, y_pred_rf))
print(confusion_matrix(y_test, y_pred_rf))
print(f"ROC-AUC Score: {roc_auc_score(y_test, y_pred_rf):.4f}")
```

```python
from sklearn.linear_model import LogisticRegression

# Training Logistic Regression
lr_model = LogisticRegression(max_iter=1000, random_state=42)
lr_model.fit(X_train_res, y_train_res)

# Predicting
y_pred_lr = lr_model.predict(X_test)

# Evaluating
print("\n✅ Logistic Regression Evaluation:")
print(classification_report(y_test, y_pred_lr))
print(confusion_matrix(y_test, y_pred_lr))
print(f"ROC-AUC Score: {roc_auc_score(y_test, y_pred_lr):.4f}")
```

```python
from lightgbm import LGBMClassifier

# Training LightGBM
lgbm_model = LGBMClassifier(random_state=42)
lgbm_model.fit(X_train_res, y_train_res)

# Predicting
y_pred_lgbm = lgbm_model.predict(X_test)

# Evaluating
print("\n✅ LightGBM Evaluation:")
print(classification_report(y_test, y_pred_lgbm))
print(confusion_matrix(y_test, y_pred_lgbm))
print(f"ROC-AUC Score: {roc_auc_score(y_test, y_pred_lgbm):.4f}")
```

```python
from catboost import CatBoostClassifier

# Training CatBoost
catboost_model = CatBoostClassifier(verbose=0, random_state=42)
catboost_model.fit(X_train_res, y_train_res)

# Predicting
y_pred_catboost = catboost_model.predict(X_test)

# Evaluating
print("\n✅ CatBoost Evaluation:")
print(classification_report(y_test, y_pred_catboost))
print(confusion_matrix(y_test, y_pred_catboost))
print(f"ROC-AUC Score: {roc_auc_score(y_test, y_pred_catboost):.4f}")
```

```python
from sklearn.ensemble import ExtraTreesClassifier

# Training Extra Trees
et_model = ExtraTreesClassifier(random_state=42)
et_model.fit(X_train_res, y_train_res)

# Predicting
y_pred_et = et_model.predict(X_test)

# Evaluating
print("\n✅ Extra Trees Evaluation:")
print(classification_report(y_test, y_pred_et))
print(confusion_matrix(y_test, y_pred_et))
print(f"ROC-AUC Score: {roc_auc_score(y_test, y_pred_et):.4f}")
```

```python
import pickle

# Saving the final model
with open('../models/fraud_detection_xgboost_v1.pkl', 'wb') as file:
    pickle.dump(xgb_model, file)

print("✅ Final XGBoost Model saved successfully as 'fraud_detection_xgboost_v1.pkl'!")
```
